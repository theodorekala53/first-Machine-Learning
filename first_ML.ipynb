```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import dataset
dataset = pd.read_csv("C://Users//Rassa//Desktop//WS2023//ML//Datensätzen/dataset.csv")
dataset.head()
```

```python
# now let first define x and y
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
```

```python
# lets see the value for x and y
X
```

```python
# now y
y
```

```python
# now lets split the dataset into training and testing 70, 30
# first we need to import the library for this process
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

#random_state=0 stellt durch Festlegen des Zufallsstamms sicher, dass die Aufteilung reproduzierbar ist.
```

```python
# lets see the value together
X_train, X_test, y_train, y_test
```

```python
# now lets work with Regression 
# first we need to import the library for regression
from sklearn.linear_model import LinearRegression

# lets make a function for regression as reg
reg = LinearRegression()

#lets fit the mode with our split dataset, first we try for 
# train and then for test and compare
reg.fit(X_train, y_train)
```

```python
# now lets predict, 9 value but prediction values
#always different
y_predict = reg.predict(X_test)
y_predict
```

```python
# we have 9 values 
y_test
```

```python
# now lets display a grap
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, reg.predict(X_train), color='blue')
plt.title("Linear Regression salary vs experinces", size=12)
plt.xlabel("Year of Employee", size=12)
plt.ylabel("Salaries of Employee", size=12)
plt.show()
```

```python
# now lets display a grap
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, reg.predict(X_test), color='blue')
plt.title("Linear Regression salary vs experinces", size=12)
plt.xlabel("Year of Employee", size=12)
plt.ylabel("Salaries of Employee", size=12)
plt.show()
```
